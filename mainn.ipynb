```python
def daysBetweenDates(year1, month1, day1, year2, month2, day2):
    # calculate maximum number of days
    maxDays = ((365*((year2-year1)+1)) ) if (year2-year1>0) else (366 if year2%4==0 else 365)
    print(maxDays)

    # compensate for leap year
    for i in range( year1,year2+1):
        print(i)
        if i%4==0:
            maxDays = maxDays + 1
    # dictionary for year 1
    dicto1={1:31,2:29 if year1%4==0 else 28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    print(dicto1)

    #dictionary for year 2
    dicto2={1:31,2:29 if year2%4==0 else 28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}
    print(dicto2)

    # 2 variables to calculate how many days to fro the max days
    # remove before the date 1
    daystoremovebefore=0
    # remove after the date 2
    daystoremoveafter=0
    
    # looping to sum all days according to months
    for i in range(1,month1):
        daystoremovebefore = daystoremovebefore+dicto1[i]
        print(i,daystoremovebefore)   
    for j in range(month2,13):
        daystoremoveafter = daystoremoveafter+dicto2[j]
        print(j,daystoremoveafter)   

    # day1 is added because before will include the number of days passed
    print(daystoremovebefore)    
    daystoremovebefore = daystoremovebefore + day1
    print(daystoremovebefore)

    # to include the number of days that have passed in the date 2
    print(daystoremoveafter)
    daystoremoveafter = daystoremoveafter - day2
    print(daystoremoveafter)

    # abra kadabra result is here
    daystoremovesum = daystoremovebefore + daystoremoveafter
    print(daystoremovesum)
    print(maxDays)
    print(maxDays - daystoremovesum)
    return maxDays - daystoremovesum

def testDaysBetweenDates():
    
    # test same day
    assert(daysBetweenDates(2017, 12, 30,
                             2017, 12, 30) == 0)
    # test adjacent days
    assert(daysBetweenDates(2017, 12, 30, 
                             2017, 12, 31) == 1)
    # test new year
    assert(daysBetweenDates(2017, 12, 30, 
                              2018, 1,  1)  == 2)
    # test full year difference
    assert(daysBetweenDates(2012, 6, 29,
                             2013, 6, 29)  == 365)
    
    print("Congratulations! Your daysBetweenDates")
    print("function is working correctly!")
    
testDaysBetweenDates()
```
